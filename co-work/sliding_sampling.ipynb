```python
import pandas as pd
import numpy as np
```

```python
csv_path = '/content/drive/MyDrive/Project/Project_Gotcha/ipynb/datasets/X_joint_data.csv'
```

```python
df = pd.read_csv(csv_path)
df
```

```python
count = df['index'].value_counts().to_frame().sort_index()
count
```

```python
from itertools import accumulate

list_start = [0] * len(count.index)
list_end = list(accumulate(count['index'] - 1))

print(list_start)
print(list_end)
```

```python
videos = list(set(df['index']))
videos.sort()
videos
```

```python
frames = []
for v in videos:
    frames.append(df[df['index'] == v].index)

frames[:2]
```

```python
n_frames = []
for f in frames:
    n_frames.append(f[-1] - f[0] + 1)

n_frames
```

```python
target_frame = 32

# sl = []
slices = []
for frame in frames:
        sl = []
        for f in frame:     
            if f <= frame[-1] - target_frame + 1:
                sl.append((f, f + target_frame - 1))
                
        slices.append(sl)

for s in slices:
    print(s)
```
