```python
import json
import glob
import pandas as pd
```

```python
file_list = glob.glob('/Users/gilbert/Developer/Project/3_Convergence/School_CCTV_AI/datasets/smoking/*')
file_list.sort()
file_list
```

```python
# joint 값 전처리 함수
def prepro(file_list=None):
    result = [] 
    n_frame = []
    for file_path in file_list:
        with open(file_path, "r") as json_file:
            # joint data 추출부
            jsonObject = json.load(json_file)
            jsonArray = jsonObject.get('annotations') # joint 값 추출
            lst = []
            keypoints = []
            for list in jsonArray:
                keypoint = list.get('keypoints')
                for i in range(len(keypoint)): # x,y 좌표 뒤에 위치한 확률값 제외하고 가져오기
                    if i % 3 != 2:
                        lst.append(keypoint[i])
                    else:
                        pass
                keypoints.append(lst[:32])
                lst=[]
            n_frame.append(len(keypoints))
            result.append(keypoints)                            
    return result, n_frame
```

```python
# 인덱스 list 생성 함수
def CreIndexList(file_list):    
    index_list=[]
    for file_path in file_list:
        # 인덱스 list 생성
        temp=file_path.split('/')
        temp=temp[-1][:-8]
#         index=temp.replace('Json/','')
        index_list.append(temp)
    return index_list
```
