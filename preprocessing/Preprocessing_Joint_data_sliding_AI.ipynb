# joint data sliding

```python
import pandas as pd
from tqdm import tqdm
```

## Data Load

```python
df = pd.read_csv('/Users/gilbert/Developer/Project/3_Convergence/School_CCTV_AI/datasets/running_nosl.csv')
df
```

## Sliding

```python
count = df['index'].value_counts()
count_df = pd.DataFrame(count).sort_index()

count_df
```

```python
videos = list(set(df['index']))
videos.sort()

frames = []
for v in videos:
    frames.append(df[df['index'] == v].index)
```

```python
target_frame = 32
slices = []
for frame in frames:
        sl = []
        for f in frame:     
            if f <= frame[-1] - target_frame + 1:
                sl.append((f, f + target_frame - 1))
        slices.append(sl)
```

```python
for sl in slices:
    for (start,end) in sl:
        temp = (start,end)
        st=temp[0]
        ed=temp[1]
```

```python
df1 = pd.DataFrame()

for sl in tqdm(slices):
    for (start, end) in sl:
        temdf=df.iloc[start:end+1] # 자른 df
        df1=pd.concat([df1, temdf],ignore_index=True)
```

## Resize

```python
n_frame = 32

for i in range(n_frame):
    if i % 2 == 0:
        df1[str(i)] = df1[str(i)] / 1920.0 * 640.0
    else:
        df1[str(i)] = df1[str(i)] / 1080.0 * 480.0

df1.head()
```

```python
df1['index'] = [1] * df1.shape[0]
```
