```python
y_train_df = pd.read_csv(y_path[0], names=['y'])
print(y_train_df.shape)

y_test_df = pd.read_csv(y_path[1], names=['y'])
print(y_test_df.shape)

print((y_train_df.shape[0] + y_test_df.shape[0], 1))
```

```python
a = dict(y_train_df.value_counts().sort_index())
a
```

```python
b = dict(y_test_df.value_counts().sort_index())
b
```

```python
a[(1, )]
```

```python
y_a1 = [0] * a[(1, )] * 32
y_a2 = [1] * a[(2, )] * 32
y_a3 = [2] * a[(3, )] * 32
y_a4 = [3] * a[(4, )] * 32
y_a5 = [4] * a[(5, )] * 32
y_a6 = [5] * a[(6, )] * 32

y_b1 = [0] * b[(1, )] * 32
y_b2 = [1] * b[(2, )] * 32
y_b3 = [2] * b[(3, )] * 32
y_b4 = [3] * b[(4, )] * 32
y_b5 = [4] * b[(5, )] * 32
y_b6 = [5] * b[(6, )] * 32

y = y_a1 + y_a2 + y_a3 + y_a4 + y_a5 + y_a6 + \
    y_b1 + y_b2 + y_b3 + y_b4 + y_b5 + y_b6

y_df = pd.DataFrame(y, columns=['y'])

y_df
```

```python
df = pd.concat([x_df, y_df], axis=1)
df.reset_index(drop=True, inplace=True)
df
```
